```python
from conut import MechanicalGrapheneBulk


mg = MechanicalGrapheneBulk(Îº=1, m=1, l=1, precision=1e-1)
evals, evecs = mg.forward()
```

```python
import numpy as np
evals, evecs = evals.numpy(), evecs.numpy()
# 3D dispersion
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
X, Y = np.meshgrid(mg.kxs, mg.kys)

vmax = np.amax(evals[:, :, 3].real)
ax.plot_surface(X, Y, evals[:, :, 1].real, vmin=0., vmax=vmax, cmap='gist_rainbow_r') 
ax.plot_surface(X, Y, evals[:, :, 2].real, vmin=0., vmax=vmax, cmap='gist_rainbow_r') 
p = ax.plot_surface(X, Y, evals[:, :, 3].real, vmin=0., vmax=vmax, cmap='gist_rainbow_r') 
ax.set_xlabel(r"$k_x$ [/m]", fontsize=14)
ax.set_xlim(-mg.lat.xw, mg.lat.xw)
ax.set_ylabel(r"$k_y$ [/m]", fontsize=14)
ax.set_ylim(-mg.lat.yw, mg.lat.yw)
ax.set_zlabel(r"$\omega$ [/s]", fontsize=14)
ax.set_zlim(0)

fig.colorbar(p, pad=0.15)
plt.show()



```
