```python
from mlat.dim2 import KagomeLattice


kl = KagomeLattice(k = [1., 1.], m=[1., 1., 1.], precision=0.1)
ws = kl.dispersion()
ws.shape
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Surface(z=ws[:, :, i], x=kl.qxs, y=kl.qys) for i in [3, 4]])
fig.update_layout(
    title='Dispersion relation of Kagome-Lattice', 
    scene=dict(
        xaxis_title=r'$k_x$',
        yaxis_title=r'$k_y$',
        zaxis_title=r'$\omega$'
    ),
    autosize=True)
fig.show()
```
