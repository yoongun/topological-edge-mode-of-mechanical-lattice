```python
from meclat import KagomeLattice


kl = KagomeLattice(k = [1., 1.5], m=[.9, 1., 1.1], precision=0.1)
dispersion = kl.dispersion()
```

```python
import plotly.graph_objects as go

# fig = go.Figure(data=[go.Surface(z=dispersion[:, :, i], x=ml2d.qxs, y=ml2d.qys) for i in range(4)])
fig = go.Figure(data=[
    go.Surface(z=dispersion[:, :, 5], x=kl.qxs, y=kl.qys),
    go.Surface(z=dispersion[:, :, 0], x=kl.qxs, y=kl.qys)])
fig.update_layout(
    title='Dispersion relation of Kagome-Lattice', 
    scene=dict(
        xaxis_title=r'$k_x$',
        yaxis_title=r'$k_y$',
        zaxis_title=r'$\omega$'
    ),
    autosize=True)
fig.show()
```
